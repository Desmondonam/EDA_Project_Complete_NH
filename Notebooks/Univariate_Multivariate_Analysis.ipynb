- Once the data is clean, We can proceed to perform Univariate and Multivariate analyses. Below is a Python script using libraries such as `matplotlib` and `seaborn` for visualization, which are essential for these types of analyses

```python
# Load the libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load cleaned data
path = r'C:\Users\Admin\Desktop\NextHikes\EDA_Project_Complete\Data\cleaned_data.csv'
df = pd.read_csv(path)
```

```python
# Set the aesthetic style of the plots
sns.set(style="whitegrid")
```

## Univariate Analysis

```python
# Histogram of SalePrice
plt.figure(figsize=(10, 6))
sns.histplot(df['SalePrice'], kde=True, color='blue')
plt.title('Distribution of Sale Prices')
plt.xlabel('Sale Price')
plt.ylabel('Frequency')
plt.show()
```

```python
# Boxplot for Overall Quality
plt.figure(figsize=(10, 6))
sns.boxplot(x='OverallQual', data=df)
plt.title('Box Plot of Overall Quality')
plt.xlabel('Overall Quality')
plt.show()
```

```python
# Count plot for number of bedrooms
plt.figure(figsize=(10, 6))
sns.countplot(x='BedroomAbvGr', data=df)
plt.title('Count of Bedrooms Above Grade')
plt.xlabel('Number of Bedrooms')
plt.ylabel('Count')
plt.show()
```

## Multivariate Analysis

```python
# Correlation matrix heatmap
plt.figure(figsize=(12, 10))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Correlation Matrix of Variables')
plt.show()
```

```python
# Pairplot for selected variables
selected_columns = ['SalePrice', 'OverallQual', 'GrLivArea', 'GarageCars', 'TotalBsmtSF', 'FullBath', 'YearBuilt']
sns.pairplot(df[selected_columns])
plt.title('Pairplot of Selected Variables')
plt.show()
```

```python
# Scatter plot of SalePrice vs. GrLivArea
plt.figure(figsize=(10, 6))
sns.scatterplot(x='GrLivArea', y='SalePrice', data=df)
plt.title('Sale Price vs. Ground Living Area')
plt.xlabel('Ground Living Area (sq ft)')
plt.ylabel('Sale Price')
plt.show()
```
